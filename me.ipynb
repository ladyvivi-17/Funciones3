```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10)
e = np.e
def f(x, e):
    return ((e ** x)-(e ** -x))/((e ** x)+(e ** -x))
plt.figure(figsize=(8, 5))
plt.plot(x, f(x, e), 'c')

plt.ylabel("Sigmoidal-Shaped")
plt.xlabel("Universe variable (arb)")
plt.ylim(-1.05, 1.05)


plt.show()
```
